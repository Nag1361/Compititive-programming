```python
class FenwickTree:
    def __init__(self, n):
        self.n = n
        self.tree = [0] * (n + 1)

    def update(self, index, delta):
        while index <= self.n:
            self.tree[index] += delta
            index += index & -index

    def query(self, index):
        total = 0
        while index > 0:
            total += self.tree[index]
            index -= index & -index
        return total


# Number of days
n = int(input("Number of days: "))

# Attendance points
attendance = list(map(int, input("Daily attendance points: ").split()))

fenwick = FenwickTree(n)

# Build Fenwick Tree
for i in range(n):
    fenwick.update(i + 1, attendance[i])

# Query 1: total attendance till Day 4
print("Total attendance till Day 4 =", fenwick.query(4))

# Update: Day 3 attendance from 1 → 2
day = 3
new_value = 2
delta = new_value - attendance[day - 1]
attendance[day - 1] = new_value
fenwick.update(day, delta)

# Query 2: total attendance till Day 4
print("After update, total attendance till Day 4 =", fenwick.query(4))
```
